```python
import matplotlib.pyplot as plt
%matplotlib inline

from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog, local_binary_pattern
from skimage.color import rgb2gray

import warnings
warnings.filterwarnings('ignore')
```

```python
img = imread(r'split\train_positif\dataset (50).png')
resized_img = resize(img, (128, 128, 3))
resized_img = rgb2gray(resized_img)
fd, hog_img = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                                  cells_per_block=(2, 2), visualize=True)
```

## Visualize HOG

```python
plt.figure()
plt.imshow(resized_img, cmap=plt.cm.gray)
```

```python
plt.figure()
plt.imshow(hog_img, cmap=plt.cm.gray)
```

## Visualize LBP

```python
radius = 3
# Number of neighbors to consider for LBP
n_points = 24
# Sampling type for LBP
METHOD = 'uniform'

lbp = local_binary_pattern(resized_img, n_points, radius, METHOD)
```

```python
plt.figure()
plt.imshow(lbp, cmap=plt.cm.gray)
```
